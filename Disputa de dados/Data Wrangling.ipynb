````python
# Bibliotecas utilizadas
import pandas as pd
import seaborn as sns 
import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use('seaborn')
import warnings
warnings.filterwarnings("ignore")
````

Extraia os dados baixados de https://covid.saude.gov.br/

````python
# C√≥digo para extrair os arquivos
csv1 = pd.read_csv('C:/Users/vinip/Downloads/HIST_PAINEL_COVIDBR_2020_Parte1_17ago2022.csv',sep=';', index_col=['data'])
csv2 = pd.read_csv('C:/Users/vinip/Downloads/HIST_PAINEL_COVIDBR_2020_Parte2_17ago2022.csv',sep=';', index_col=['data'])
csv3 = pd.read_csv('C:/Users/vinip/Downloads/HIST_PAINEL_COVIDBR_2021_Parte1_17ago2022.csv',sep=';', index_col=['data'])
csv4 = pd.read_csv('C:/Users/vinip/Downloads/HIST_PAINEL_COVIDBR_2021_Parte2_17ago2022.csv',sep=';', index_col=['data'])
csv5 = pd.read_csv('C:/Users/vinip/Downloads/HIST_PAINEL_COVIDBR_2022_Parte1_17ago2022.csv',sep=';', index_col=['data'])
csv6 = pd.read_csv('C:/Users/vinip/Downloads/HIST_PAINEL_COVIDBR_2022_Parte2_17ago2022.csv',sep=';', index_col=['data'])
````

Fa√ßa aqui o concatenamento das linhas dos `dataframes` da lista em um `dataframe` √∫nico. 

````python
arquivos = pd.concat([csv1, csv2, csv3, csv4, csv5, csv6], axis=0)
arquivos
````

Nosso objetivo √© mostrar somente os valores a n√≠vel de Brasil. Fa√ßa um filtro das linhas baseado na coluna `regiao` somente para a correspond√™ncia `Brasil`. Salve este filtro em um `Dataframe`.

````python
csv_BR = arquivos[arquivos['regiao'] == 'Brasil']
csv_BR
````

Selecione somente as colunas 'casosAcumulado', 'casosNovos', 'obitosAcumulado', 'obitosNovos' do `Dataframe` e descarte as demais.

````python
csv_BR = csv_BR[['casosAcumulado', 'casosNovos', 'obitosAcumulado', 'obitosNovos']]
csv_BR
````

Fa√ßa aqui o gr√°fico de 'casosAcumulado' ao longo dos dias desde o come√ßo da pandemia.

````python
csv_BR['casosAcumulado'].plot(color='firebrick', linewidth=10, figsize=(60,30))

plt.xticks(fontsize=50)
plt.yticks(fontsize=50)
plt.legend('')

plt.title('Casos Acumulados ao longo dos dias desde o come√ßo da pandemia', fontsize=70)
plt.xlabel('Data', fontsize=70)
plt.ylabel('Casos Acumulados', fontsize=70)
````

Fa√ßa aqui o gr√°fico de 'obitosAcumulado' ao longo dos dias desde o come√ßo da pandemia.

````python
csv_BR['obitosAcumulado'].plot(color='firebrick', linewidth=10, figsize=(60,30))

plt.xticks(fontsize=50)
plt.yticks(fontsize=50)
plt.legend('')

plt.title('Obitos Acumulados ao longo dos dias desde o come√ßo da pandemia', fontsize=70)
plt.xlabel('Data', fontsize=70)
plt.ylabel('Obitos Acumulados', fontsize=70)
````

Crie uma coluna com a media m√≥vel dos 'casosNovos' e 'obitosNovos'. Use uma janela de 7 dias para o c√°lculo da m√©dia m√≥vel.

Use o seguinte exemplo como refer√™ncia para calcular a m√©dia m√≥vel.

```python

df['nomeColuna'].rolling(Janela).mean()
```

````python
csv_BR['media_movel_casosNovos'] = csv_BR['casosNovos'].rolling(7).mean()
csv_BR
````

````python
csv_BR['media_movel_obitosNovos'] = csv_BR['obitosNovos'].rolling(7).mean()
csv_BR
````

Fa√ßa um gr√°fico da evolu√ß√£o dos casos e sua m√©dia m√≥vel. Use a coluna 'casosNovos' para a evolu√ß√£o di√°ria de casos e a coluna referente √† m√©dia m√≥vel de casos que voc√™ calculou.

````python
csv_BR['casosNovos'].plot(color='black', linewidth=10, figsize=(60,30))
csv_BR['media_movel_casosNovos'].plot(color='lime', linewidth=7)

plt.xticks(fontsize=50)
plt.yticks(fontsize=50)
plt.legend('')

plt.title('Evolu√ß√£o dos Casos de COVID-19 com a M√©dia M√≥vel', fontsize=20)
plt.xlabel('Data', fontsize=70)
plt.ylabel('Casos Novos', fontsize=70)
````

Fa√ßa um gr√°fico da evolu√ß√£o dos √≥bitos e sua m√©dia m√≥vel. Use a coluna 'obitosNovos' para a evolu√ß√£o di√°ria de √≥bitos e a coluna referente √† m√©dia m√≥vel de √≥bitos que voc√™ calculou.

````python
csv_BR['obitosNovos'].plot(color='black', linewidth=10, figsize=(60,30))
csv_BR['media_movel_obitosNovos'].plot(color='lime', linewidth=7)

plt.xticks(fontsize=50)
plt.yticks(fontsize=50)
plt.legend('')

plt.title('Evolu√ß√£o dos Obitos de COVID-19 com a M√©dia M√≥vel', fontsize=20)
plt.xlabel('Data', fontsize=70)
plt.ylabel('Obitos Novos', fontsize=70)
````
